```python
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import save_img
import os
import cv2
import json
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import random
```

```python
good_train_data_dir = "/home/jupyter/large-dataset/train-data-large/Good"
bad_train_data_dir = "/home/jupyter/large-dataset/train-data-large/Bad"
good_validation_data_dir = "/home/jupyter/large-dataset/validation-data-large/Good"
bad_validation_data_dir = "/home/jupyter/large-dataset/validation-data-large/Bad"
good_paths = []
for image in os.listdir(good_train_data_dir):
    good_paths.append( os.path.sep.join([good_train_data_dir, image]))
for image in os.listdir(good_validation_data_dir):
    good_paths.append( os.path.sep.join([good_validation_data_dir, image]))   
random.shuffle(good_paths)
```

```python
good_images = good_paths[:120]
print(len(good_images))
```

```python
labels = []
for i in range(0,120):
    labels.append('Good')
print(len(labels))
```

```python
bad_paths = []
for image in os.listdir(bad_train_data_dir):
    bad_paths.append( os.path.sep.join([bad_train_data_dir, image]))
for image in os.listdir(bad_validation_data_dir):
    bad_paths.append( os.path.sep.join([bad_validation_data_dir, image]))   
random.shuffle(bad_paths)
bad_images = bad_paths[:44]
for i in range(0,44):
    labels.append('Bad')
print(len(bad_images))
```

```python
train_ratio=0.80
imagePaths = good_images + bad_images
x_train, x_val, y_train, y_val = train_test_split(imagePaths, labels, test_size=1 - train_ratio)
train_path = "/home/jupyter/large-dataset/small-dataset/training"
validation_path = "/home/jupyter/large-dataset/small-dataset/validation"
i = 0
while i < len(x_train) :
    img = load_img(x_train[i])
    img_array = img_to_array(img)
    print(x_train[i][-38:])
    print(y_train[i])
    save_img(os.path.sep.join([train_path, y_train[i], x_train[i][-38:]]), img_array)
    i += 1
```

```python
print(len(os.listdir("/home/jupyter/large-dataset/small-dataset/training/Bad")))
print(len(os.listdir("/home/jupyter/large-dataset/small-dataset/training/Good")))
```

```python
i = 0
while i < len(x_val) :
    img = load_img(x_val[i])
    img_array = img_to_array(img)
    save_img(os.path.sep.join([validation_path, y_val[i], x_val[i][-38:]]), img_array)
    i += 1
```

```python
print(len(os.listdir("/home/jupyter/large-dataset/small-dataset/validation/Bad")))
print(len(os.listdir("/home/jupyter/large-dataset/small-dataset/validation/Good")))
```
