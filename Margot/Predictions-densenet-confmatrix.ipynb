```python
import os
import tensorflow as tf
import keras
from keras.applications import inception_v3 as inc_net
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions
from skimage.io import imread
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
medium_model = tf.keras.models.load_model('/home/jupyter/Cropping-tuner/DenseNet-MediumDataModel')
```

```python
small_model = model = tf.keras.models.load_model('/home/jupyter/Cropping-tuner/DenseNet-SmallDataModel')
```

```python
for image_path in os.listdir("/home/jupyter/large-dataset/test-data-large/Bad"):
    img = image.load_img(os.path.join("/home/jupyter/large-dataset/test-data-large/Bad", image_path), target_size=(224, 224))
    img = image.img_to_array(img)/255
    prediction_scores = medium_model.predict(np.expand_dims(img, axis=0))
    predicted_index = (prediction_scores[0][0] > 0.5).astype(np.int)
    print("Predicted label: " +str(predicted_index))
```
