```python
#Google streetview query 
#tutorial: https://medium.com/data-mining-the-city/sorting-google-street-view-68665f8fa8e8
#code from https://github.com/awmartin/spatialpixel/blob/master/Sketches/Intro/SimpleGoogleStreetView2/SimpleGoogleStreetView2.pyde
import csv
import urllib
import pandas as pd
from google.cloud import storage
from datetime import datetime
```

```python
data= pd.read_csv("/home/jupyter/leuvenAddresses.csv", sep=",", low_memory=False)
```

```python
data.head()
```

```python
locations=data.values
print(locations[0])
print(locations[0][1])
addresses = data.iloc[:,1].str.replace(" ","") + data.iloc[:,0] + ',' + data.iloc[:,2]
print(addresses[0])
```

```python
# Put your Google Street View API key here.
api_key = ""

api_url = "https://maps.googleapis.com/maps/api/streetview?size=600x600&location={0}&fov=90&pitch=10&source=outdoor&key={1}"


```

```python
client = storage.Client()
client = storage.Client(project="fednot-sandbox-KUL")
bucket = client.get_bucket("fednot-sandbox-kul-streetview")
```

```python
print(datetime.now())
streets=["Martelarenlaan", "Marterlarenplein","Aarschotsesteenweg", "Naamsesteenweg", "Riddersstraat", "Roeselbergdal", "Rotselaarsesteenweg", "Schorenshof", "Sint-Jansbergsesteenweg", "Steenakkerstraat", "Tervuursesteenweg", "Tiensesteenweg", "Wijgmaalsesteenweg", "Wilselsesteenweg", "Brusselsesteenweg", "Diestsesteenweg", "Diestsestraat", "ErasmeRuelenspark", "Holsbeeksesteenweg", "JefVanHooflaan", "Kerspelstraat", "Kessel-Losesteenweg", "KorteHolsbeeksesteenweg", "Kriekenboslaan", "Leming", "Lentedreef", "Schepenstraat", "Mechelsesteenweg"]
row=0
count=0
# Loop over every location, and for each location, loop over all the possible headings.
#images4: ]1299,2314[
#images5: ]2314, 3330[
#images6: ]3330, 4464[
#images7: ]4464, 6713[
#images8: ]6713, 10181[
#images9: ]10181, 11882[
#images10: ]11882, 13600[
#images11: ]13600,15021[ & ]15021,17132[
for location in addresses:
    if row <= 15021:
        row +=1
        continue
    #brabançonnestraat op row 43 geeft foutmelding, is rij 44 in dataset
    #Léon schreursvest
    if ("ç" in location) or ("é" in location) or ("è" in location) or ("à" in location):
        row +=1
        continue
    badstreet = False
    for street in streets: 
        if (street in location):
            badstreet=True
            break
    if badstreet:
        row +=1
        continue
    if count == 1800:
        break
    # Create the URL for the request to Google.
    url = api_url.format(location, api_key)

    # Create the filename we want to save as, e.g. location-2-90.jpg.
    filename = "{0}.jpg".format(location.replace(",Leuven", ""))

    # Use the 'curl' command to actually make the request and save the file to disk.
    urllib.request.urlretrieve(url, filename)
    #print "Got %s" % (filename,)
    #blob = storage.Blob(filename, bucket)
    #blob.upload_from_filename(filename)
        
    print(filename) 
    count +=1
    # Increment the row to correlate with the Excel file.
    row += 1
   
print("Done: "+str(count))
print(row)
print(datetime.now())
```
